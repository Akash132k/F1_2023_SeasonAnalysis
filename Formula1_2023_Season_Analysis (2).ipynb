## 2. 2023 Sprint Race Analysis

Before sprint races were introduced the format was practice sessions on friday, qualifying seeions on saturday and the primary race on sunday.

Sprint races are the new additions of F1 race weekend structure, first introduced in 2021 Season. In such weekends where this kind of race is available, qualifying session is carried out on Friday to determine the starting grid of the sprint race that is implemented on Saturday and also the award for pole position.

Note that the distances of sprint races are roughly one third of a Sunday's race session (in other words, a little more than 100 km) and it's not mandatory to pit in for changing tyres for another compound. The main purpose of these races are to shuffle the grid a bit more by racing on the track to prepare all drivers and teams for the real Grand Prix race on Sunday. Moreover, the top 8 drivers can earn additional championship points, where the winner of the sprint race gets 8 points, going down one by one to 1 point for the driver finishing eighth, instead of just top 3 finishers in 2021 Season.

6 Sprint races were held in the 2023 F1 season and **Max Verstappen** unsurprisingly won 4 of them. It was refreshing to see McLaren's rookie **Oscar Piastri** win his first ever sprint race in the Qatar Gp. He won the silverware before his significantly experienced teammate **Lando Norris** could do.

### 5.Points Progression of the top 10 drivers of the season

Let us take a look at how the top 10 drivers progressed through all the races of the 2023 season:

```python
Top10DriverStandings = driverStandings['Driver'][:10].values
Top10DriverPoints = {}
Top10DriverPointsSprint = {}
for driver in Top10DriverStandings:
    Top10DriverPoints[driver] = f1_2023results[f1_2023results['Driver'] == driver]['Points'].values
    Top10DriverPointsSprint[driver] = f1_2023Sprintresults[f1_2023Sprintresults['Driver'] == driver]['Points'].values

SprintRounds = [3, 9, 12, 17, 18, 20]
for driver in Top10DriverStandings:
    for i in range(len(SprintRounds)):
        Top10DriverPoints[driver][SprintRounds[i]] += Top10DriverPointsSprint[driver][i]
```

```python
tracks = f1_2023results['Track'].unique()

colours = ['#0600f0', '#0600f0', '#00d2bd', 'SeaGreen', 'red', 'orange', 'red', '#00d2bd', 'orange', 'SeaGreen']

plt.figure(figsize=(12, 6))

for i in range(10):
    if Top10DriverStandings[i] in ['Sergio Perez', 'Carlos Sainz', 'George Russell', 'Lance Stroll','Oscar Piastri']: 
        ls = '--'
    else: 
        ls = '-'
    plt.plot(Top10DriverPoints[Top10DriverStandings[i]].cumsum(), label=Top10DriverStandings[i].split()[1], 
             c=colours[i], linewidth=2, ls=ls, zorder = 4)

plt.title('F1- 2023 Season\nTop 10 Drivers\' Points Progression', fontsize=19, fontweight='bold')
plt.xlabel('Races', fontsize=16, fontweight='bold',)
plt.ylabel('Points', fontsize=16, fontweight='bold')
plt.xticks(range(0, len(tracks)), tracks, rotation=60, fontsize=10)
plt.yticks(range(0, 600, 50), fontsize=12,)
plt.grid(axis='y', color='gray', zorder=1)
plt.grid(axis='x', color='gray', zorder=1)
plt.legend()
plt.show()

```

### 9. 2023 DNFs

The upcoming section of the analysis will focus on DNFs (Did Not Finish). It can be considered unfortunate for any driver to retire from a race due to various reasons, including self-driving errors resulting in collisions with barriers or getting stuck in sandpits, sustaining performance-affecting damage from collisions with other competitors, or experiencing mechanical failures such as engine or transmission issues.

The table and the bar chart shows how many times a driver did not finish a race:

```python
DNFdriver = f1_2023results[f1_2023results['Time/Retired'] == 'DNF'].groupby(['Driver','Team'])['Driver'].value_counts().sort_values(ascending=False)
DNFs = pd.DataFrame(DNFdriver).fillna(0).reset_index()
DNFs['S.No']=range(1,20)
DNFs.set_index('S.No', inplace=True)
DNFs
```

```python
DNFdriver = f1_2023results[f1_2023results['Time/Retired'] == 'DNF']['Driver'].value_counts()

plt.figure(figsize=(11, 5))
plt.barh(DNFdriver.index, DNFdriver,color='Tomato', height = 0.5, zorder = 4)

for index, value in enumerate(DNFdriver):
    plt.text(value, index, str(value),fontweight='bold')

plt.xlabel('No of DNFs',fontsize=14, fontweight='bold')
plt.ylabel('Driver',fontsize=14, fontweight='bold')
plt.title('Number of DNF by drivers',fontsize=14, fontweight='bold')
plt.grid(axis='y', color='gray', zorder=1)
plt.grid(axis='x', color='gray', zorder=1)
plt.gca().invert_yaxis()  
plt.show()
```

Except for **Max Verstappen**, **Liam Lawson**, **Daniel Ricciardo**, every other driver was a victim of DNF.

### 10. Driver of the day votes

While no points are awarded to the Driver of the Day, it's a notable accolade for the recipient, highlighting exceptional driving, remarkable overtakes, or simply memorable performances.
